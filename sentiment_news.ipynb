```python
import pandas as pd

# 假设news_data已经包含了'time', 'polarity', 'subjectivity'列
# 先将时间转换为日期
news_data['date'] = pd.to_datetime(news_data['date']).dt.date

# 定义一个函数来计算加权平均
def weighted_average(group):
    weighted_polarity = (group['polarity'] * group['subjectivity']).sum() / group['subjectivity'].sum()
    return pd.Series({'polarity': weighted_polarity})

# 按日期分组并应用加权平均
daily_polarity = news_data.groupby('date').apply(weighted_average).reset_index()

#保存到本地
daily_polarity.to_csv("data/daily_polarity_news.csv")
```

```python
#研究前半年
daily_polarity = daily_polarity[(daily_polarity['date'] >= pd.to_datetime('2022-01-01').date()) & (daily_polarity['date'] <= pd.to_datetime('2022-07-01').date())]
```

```python
import matplotlib.pyplot as plt

# 创建图表
plt.figure(figsize=(10, 6))

# 绘制加权平均polarity的折线图
plt.plot(daily_polarity['date'], daily_polarity['polarity'], color='blue', marker='o', alpha=0.7)

# 设置标题和标签
plt.title('Daily Weighted Average Polarity over Time')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Weighted Polarity')

# 添加网格以便更好地观察数据
plt.grid(True)

# 显示图表
plt.show()
```

```python
# 3. 合并BTC数据和每日加权平均polarity
merged_data = pd.merge(daily_polarity, BTC_2022, on='date', how='inner')

# 4. 绘制双Y轴图
fig, ax1 = plt.subplots(figsize=(12, 6))

# 绘制BTC价格的折线图
ax1.set_xlabel('Date')
ax1.set_ylabel('BTC Price', color='tab:blue')
ax1.plot(merged_data['date'], merged_data['close'], color='tab:blue', label='BTC Price')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 绘制加权平均polarity的折线图
ax2 = ax1.twinx()
ax2.set_ylabel('Weighted Polarity', color='tab:orange')
ax2.plot(merged_data['date'], merged_data['polarity'], color='tab:orange', label='Polarity', linestyle='--')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# 添加volume到下方的子图
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # 设置第三个Y轴的位置
ax3.set_ylabel('Volume', color='tab:green')
ax3.plot(merged_data['date'], merged_data['volume'], color='tab:green', alpha=0.3, label='Volume')
ax3.tick_params(axis='y', labelcolor='tab:green')

# 设置标题和显示图表
plt.title('BTC Price, Volume, and Weighted Polarity over Time')
fig.tight_layout()
plt.show()
```

```python
# 计算Pearson相关系数
correlation = merged_data['polarity'].corr(merged_data['close'])
print(f'Pearson correlation between polarity and price: {correlation:.2f}')

# 绘制散点图
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['polarity'], merged_data['close'], alpha=0.5)
plt.title('Scatter Plot of Polarity vs. Price')
plt.xlabel('Polarity')
plt.ylabel('BTC Price')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设 merged_data 已经包含 'date', 'polarity', 'price' 列

# 创建一个数据框来存储不同滞后期的相关性
lagged_correlations = {}

# 最大滞后期设为30天
max_lag = 30

# 计算每个滞后期的相关性
for lag in range(1, max_lag + 1):
    # 创建滞后变量
    merged_data[f'polarity_lag_{lag}'] = merged_data['polarity'].shift(lag)
    
    # 计算滞后相关性
    correlation = merged_data[f'polarity_lag_{lag}'].corr(merged_data['close'])
    
    # 存储相关性
    lagged_correlations[lag] = correlation

# 将结果转换为DataFrame以便绘图
lagged_correlation_df = pd.DataFrame.from_dict(lagged_correlations, orient='index', columns=['Correlation'])

# 绘制滞后相关性图表
plt.figure(figsize=(10, 6))
plt.plot(lagged_correlation_df.index, lagged_correlation_df['Correlation'], marker='o', linestyle='-')
plt.title('Lagged Correlation between Polarity and BTC Price')
plt.xlabel('Lag (days)')
plt.ylabel('Pearson Correlation')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 假设 merged_data 已经包含 'date', 'polarity', 'price' 列

# 设置滚动窗口大小（例如30天）和滞后期（例如7天）
window_size = 50
lag = 2

# 创建一个DataFrame来存储滚动相关性
rolling_correlations = []

for i in range(len(merged_data) - window_size - lag):
    window_data = merged_data.iloc[i:i + window_size + lag]
    window_data['lagged_polarity'] = window_data['polarity'].shift(lag)
    correlation = window_data['lagged_polarity'].corr(window_data['close'])
    rolling_correlations.append(correlation)

# 转换为DataFrame并绘制图表
rolling_correlation_df = pd.DataFrame(rolling_correlations, columns=['Correlation'])
rolling_correlation_df['date'] = merged_data['date'].iloc[window_size + lag:].values

plt.figure(figsize=(10, 6))
plt.plot(rolling_correlation_df['date'], rolling_correlation_df['Correlation'], marker='o', linestyle='-')
plt.title('Rolling Window Correlation between Polarity and BTC Price')
plt.xlabel('Date')
plt.ylabel('Pearson Correlation')
plt.grid(True)
plt.show()
```

```python
# 转换为DataFrame以便计算平均值
rolling_correlation_df = pd.DataFrame(rolling_correlations, columns=['Correlation'])
positive_correlations = rolling_correlation_df[rolling_correlation_df['Correlation'] > 0]
# 计算相关性的平均值
average_correlation = positive_correlations['Correlation'].mean()

print(f'Average Pearson correlation between polarity and price with lag={lag} over rolling windows: {average_correlation:.4f}')
```
